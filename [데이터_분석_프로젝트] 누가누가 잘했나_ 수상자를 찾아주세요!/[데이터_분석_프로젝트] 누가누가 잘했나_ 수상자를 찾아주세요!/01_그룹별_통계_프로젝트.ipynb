### 1) 계주 경기 수상자

> * 수상계획: 1등, 2등 팀에게 상품 지급
>
> 계주 1등 팀을 찾기 위해 계주경기의 단체승점 최고점을 확인해 봅시다. 

````python
계주_최고승점 = df.loc[df.단체경기=="계주"].단체승점.max()

print('계주경기 최고승점: ', 계주_최고승점)
````

> 최고승점은 3점입니다.  
> 계주경기의 단체승점이 3점인 팀을 찾아봅시다. 

````python
계주_1등팀 = df.loc[(df.단체경기=="계주")&(df.단체승점==3)]

print(계주_1등팀['반'].unique())
````

> 나 팀에게 계주 1등 상품을 지급하겠습니다!

> 2등의 승점을 찾아봅시다. 

````python
계주_2등승점 = df.loc[(df.단체경기=="계주")&(df.단체승점 < 3)].단체승점.max()

print(계주_2등승점)
````

> 계주경기 2등의 승점은 2점입니다.  
> 계주경기의 단체승점이 2점인 팀을 찾아봅시다. 


````python
계주_2등팀 = df.loc[(df.단체경기=="계주")&(df.단체승점==2)]

print(계주_2등팀['반'].unique())
````

> 다 팀에게 계주 2등 상품을 수여하겠습니다!

### **5) 림보 경기 수상자 찾기**

> * 림보: 학년 별 1등에게 상품 지급
> * 림보의 경우 1등이 여러명일 수 있다.
>
> for문을 이용하여 성별로 각 학년의 1등 학생(들)을 찾아봅시다.  
> 1등 학생들의 학년, 반, 번호, 성별, 개인경기, 개인승점을 출력해 봅시다.

````python
우승자_리스트 = []

for i in [1, 2, 3]:
    for s in ['남', '여']:
        
        림보_선수 = df.loc[(df.개인경기=='림보')&(df.성별==s)&(df.학년==i)]
        
        림보_최고레벨 = 림보_선수.개인승점.max()
        
        림보_우승자 = 림보_선수.loc[df.개인승점==림보_최고레벨]
        
        우승자_리스트 = 우승자_리스트 + 림보_우승자.index.tolist()
        
        
print(df.loc[우승자_리스트, ["학년", "반", "번호", "성별", "개인경기", "개인승점"]])
````
